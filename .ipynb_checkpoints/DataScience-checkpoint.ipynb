## Now Working with the Weather File

```python
weather = pd.read_csv('seattleWeather_1948-2017.csv')
```

```python
weather.head()
```

```python
weather.loc[weather['DATE'] >= '2011'].head()
```

```python
weather2011 = weather.loc[weather['DATE'] >= '2011']
```

```python
weather2011['Year'] = weather2011.DATE.astype(str).str[2:4]
```

```python
weather2011['Month'] = weather2011.DATE.astype(str).str[5:7]
```

```python
weather2011.head()
```

```python
weather2011.to_csv('weather2011.csv')
```

```python
weather2011.groupby(['Month', 'Year']).sum()
```

```python
weather2011['RAIN'] = weather2011['RAIN'].astype(int)
```

```python
weatherGrouped = weather2011.groupby(['Month', 'Year']).sum()
```

```python
weatherGrouped.index
```

```python
months = [a for a in range(1, 13)]
years = [a for a in range(2011, 2018)]
daysPerMonth = {
    1: 31,
    2: 28,
    3: 31,
    4: 30,
    5: 31,
    6: 30, 
    7: 31,
    8: 31,
    9: 30,
    10: 31,
    11: 30,
    12: 31
}
```

```python
months
```

```python
weather2011.Month.unique()
```

```python
from collections import defaultdict
yearMonthCounts = {
    11: defaultdict(int),
    12: defaultdict(int),
    13: defaultdict(int),
    14: defaultdict(int),
    15: defaultdict(int),
    16: defaultdict(int),
    17: defaultdict(int),
}

for index, row in weather2011.iterrows():
#     print(row['Year'])
#     print(row['Month'])
    if row['RAIN'] == 1:
        yearMonthCounts[int(row['Year'])][int(row['Month'])] += 1
```

```python
yearMonthCounts
```

```python
rainyMonths = {
    11: {},
    12: {},
    13: {},
    14: {},
    15: {},
    16: {},
    17: {},
}

for year, month in yearMonthCounts.items():
    for month_number, rainy_days in month.items():
        if rainy_days > float(daysPerMonth[month_number]) / 2:
            rainyMonths[year][month_number] = True
        else:
            rainyMonths[year][month_number] = False
```

```python
rainyMonths
```
