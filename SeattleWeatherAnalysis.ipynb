For this, we will use Hourly Dry Bulb Temperature for simplicity, since this is the closest thing to a standard temperature

```python
tempdf = weather_d[pd.notnull(weather_d['HourlyDryBulbTemperature'])]
```

It looks like there is infact 100% coverage of the dry bulb temperature

```python
#What month on average is the hottest!
tempdf.index = pd.to_datetime(tempdf['DATE'])
```

```python
import seaborn as sns; 
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
```

```python
fig, ax = plt.subplots(figsize = (10,10))

temprf = tempdf[:50]
# add the x-axis and the y-axis to the plot
ax.plot(temprf['DATE'], 
        temprf['HourlyDryBulbTemperature'], 
        color = 'red')

# rotate tick labels
plt.setp(ax.get_xticklabels(), rotation=45)

# set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Temperature",
       title="Temperature \n Seattle");
```

```python
temperature_data = tempdf[['HourlyDryBulbTemperature']]
```

```python
temperature_data.head()
```

```python
plt.plot_date(temperature_data.index, temperature_data['HourlyDryBulbTemperature'])
```

```python
fig, ax = plt.subplots()
ax.plot(temperature_data.index,temperature_data['HourlyDryBulbTemperature'])
```

```python
temperature_data['HourlyDryBulbTemperature'] = temperature_data.apply(pd.to_numeric, errors='coerce')
```

```python
temp_data_monthly = temperature_data.resample('M').mean()
```

```python
temp_data_monthly.head()
```

```python
temp_data_monthly.tail()
```

```python
fig, ax = plt.subplots()
ax.plot(temp_data_monthly, marker='', linestyle='-', linewidth=0.5)
ax.set_ylabel('Temperature')
ax.set_xlabel('Monthly/Yearly Time')
```

Now for precipitaiton/rain

```python
tempdf.head(10)
```

```python
precipitation_data = tempdf[['HourlyPrecipitation']].fillna(0)
```

```python
precipitation_data.head()
```

```python
precipitation_data['HourlyPrecipitation'] = precipitation_data.apply(pd.to_numeric, errors='coerce')
```

```python
prec_data_monthly = precipitation_data.resample('M').sum()
```

```python
prec_data_monthly.head()
```

```python
fig, ax = plt.subplots()
ax.plot(prec_data_monthly, marker='', linestyle='-', linewidth=0.5)
ax.set_ylabel('Precipitation Monthly Sum')
ax.set_xlabel('Monthly/Yearly Time')
```
