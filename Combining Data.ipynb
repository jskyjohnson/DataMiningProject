```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import seaborn as sns; 
from scipy import stats
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
sns.set(rc={'figure.figsize':(11, 4)})
```

```python
CheckoutData = pd.read_pickle("Data/CheckoutData.pkl")['2005-04-13 08:00:00':'2019-07-22 12:00:00'].between_time('8:00', '18:00')
display(CheckoutData)
```

```python
WeatherData = pd.read_pickle("Data/WeatherData_Unrefined.pkl").fillna(0)
display(WeatherData)
```

```python
WeatherData.convert_objects(convert_numeric=True)
```

```python
WeatherDataRefined = WeatherData.apply(lambda col:pd.to_numeric(col, errors='coerce'))
```

```python
WeatherDataRefined.dtypes
```

```python
Weather_R = WeatherDataRefined.resample('H').mean()
```

```python
WeatherData_R = Weather_R['2005-04-13 08:00:00':'2019-07-22 12:00:00'].between_time('8:00', '18:00')
```

```python
display(WeatherData_R)
display(CheckoutData)
```

```python
WeatherData_R['HourlyDryBulbTemperature'].corr(CheckoutData['ItemType'])
```

Since there's naturally a shift in participation to the library throughout the day, this correlation doesn't particularly tell us much. Instead it might be more useful to look at correlation between a specific hour and season.

```python
WeatherData_R_Copy = WeatherData_R.copy()
CheckoutData_Copy = CheckoutData.copy()
```

```python
df = WeatherData_R_Copy.add(CheckoutData_Copy, fill_value=0)
```

```python
df
```
